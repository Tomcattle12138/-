```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('/home/yzy/storage/data.csv')
X = dataset.iloc[:, 1:].values

# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, max_iter = 300, n_init = 10, init = 'k-means++', random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)# 这里是算出k-means的组间距离之和
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()



# Applying the k-means to the mall dataset
kmeans = KMeans(n_clusters = 4, max_iter = 300, n_init = 10, init = 'k-means++', random_state = 0)
y_kmeans = kmeans.fit_predict(X)
dataset["cluster"] = y_kmeans
dataset.to_csv("/home/yzy/storage/data.csv",index=False)
new_data = pd.read_csv("/home/yzy/storage/data.csv")
print(new_data)
print(y_kmeans)
print(len(X[y_kmeans==0,0]))
print(len(X[y_kmeans==1,0]))
# Visualizing the clusters
plt.scatter(X[y_kmeans == 0, 0], [1]*len(X[y_kmeans == 0, 0]), s = 1, c = 'red', label = 'cluster 1')
plt.scatter(X[y_kmeans == 1, 0], [1]*len(X[y_kmeans == 1, 0]), s = 1, c = 'blue', label = 'cluster 2')
plt.scatter(X[y_kmeans == 2, 0], [1]*len(X[y_kmeans == 2, 0]), s = 1, c = 'yellow', label = 'cluster 3')
plt.scatter(X[y_kmeans == 3, 0], [1]*len(X[y_kmeans == 3, 0]), s = 1, c = 'black', label = 'cluster 4')
plt.title('Clusters of clients')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
